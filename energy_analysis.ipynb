```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

energy = pd.read_csv("data_files/processed/energy_cleaned.csv")
events = pd.read_csv("data_files/processed/events_cleaned.csv")
```

```python
#energy distribution from 2008-2018 data
fig1, ax1 = plt.subplots(figsize = (8,5))
sns.histplot(energy["log10(E/GeV)"], bins=30, kde=True, color='lightgreen', edgecolor='red')

ax1.set_title("iceCube Neutrino Energy Distribution")
ax1.set_xlabel("log10(E/GeV)")
ax1.set_ylabel("count")
ax1.grid(alpha=0.3)
plt.show()

fig1.savefig("figures/energy_dist.png", dpi=300, bbox_inches="tight")
```

```python
#energy distribution from 2008-2018
fig2, ax2 = plt.subplots(figsize=(8,5))
sns.histplot(
    data=energy[energy["direction"] == "up-going"],
    x="log10(E/GeV)",
    bins=25,
    kde=True,
    color='lightgreen',
    alpha=0.6,
    label='up-going'
)

sns.histplot(
    data=energy[energy["direction"] == "down-going"],
    x='log10(E/GeV)',
    bins=25,
    kde=True,
    color="orange",
    alpha=0.6,
    label="down-going"
)
ax2.set_title('IceCube Neutrino Energy: Up-going vs. Down-going')
ax2.set_xlabel('log10(E/GeV)')
ax2.set_ylabel('count')
ax2.grid(alpha=0.3)
ax2.legend(title = 'neutrino direction' )

plt.show()
fig2.savefig("figures/energy_direction.png", dpi=300, bbox_inches="tight")
```

```python
#scatterplot of direction and energy 
fig3, ax3 = plt.subplots(figsize=(8,5))

ax3.hist2d(
    energy["Zenith[deg]"],
    energy["log10(E/GeV)"],
    bins=60,
    cmap="plasma"
)

ax3.set_title("Neutrino Energy vs Zenith Angle (Density Map Form)")
ax3.set_xlabel("zenith angle (deg)")
ax3.set_ylabel("log10(E/GeV)")
ax3.grid(alpha = 0.3)

plt.show()

fig3.savefig("figures/zenith_angle_density.png", dpi=300, bbox_inches="tight")
```
