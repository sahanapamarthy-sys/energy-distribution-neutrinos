```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

energy = pd.read_csv("data_files/processed/energy_cleaned.csv")
events = pd.read_csv("data_files/processed/events_cleaned.csv")
```

```python
#distribution of energy vs event types 
fig4, ax4 = plt.subplots(figsize=(9,6))
sns.violinplot(
    data=events,
    x="reconstruction",
    y="energy",
    ax=ax4,
    inner="quartile",
    density_norm="width", 
    linewidth=1.5
)

ax4.set_title("energy distribution by event type (Track vs. Shower)")
ax4.set_xlabel("reconstruction type")
ax4.set_ylabel("log10(E/GeV)")
ax4.grid(alpha=0.3)
plt.show()


fig4.savefig("figures/energy_event.png", dpi=300, bbox_inches="tight")
```
