```python
#importing relevant libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#loading data from local repo
with open("data_files/rawdata/neutrino2008.csv", "r") as f:
    header = f.readline().replace("#", "").strip().split()
    
datax = pd.read_csv("data_files/rawdata/neutrino2008.csv", sep=r"\s+", skiprows = 1, names = header)
datay = pd.read_csv("data_files/rawdata/events.csv" )
```

```python
print(datax.head())
print(datay.head())
```

```python
# so i am making the following graphs: 
# energy distribution of neutrinos (number of neutrinos vs energy)
# up going vs down going energy distributions (split historgram scatter plot energy vs zenith)
```

```python
print(datax.columns)
print(datay.columns)
```

```python
#keeping relevant columns (cleaning)
energy = datax[['log10(E/GeV)', 'Zenith[deg]']]
events = datay[['energy', 'reconstruction']]

energy = energy.dropna()
events = events.dropna()

```

```python
#categorizing direction 
energy["direction"] = np.where(energy['Zenith[deg]'] > 90, "up-going", "down-going")
print(energy.head())
print(energy.describe())
print(energy["direction"].value_counts())
```

```python
#categorizing types of neutrinos 
print(events['reconstruction'].value_counts())
print(events.head())
```

```python
energy.to_csv("data_files/processed/energy_cleaned.csv", index=False)
events.to_csv("data_files/processed/events_cleaned.csv", index=False)
```
